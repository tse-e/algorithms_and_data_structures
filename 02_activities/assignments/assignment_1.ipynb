
-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
# Dictionary to define the accepted bracket pairs
matches = {")": "(", "]": "[", "}": "{"}

def is_valid_brackets(s): 
    opening_brackets = []  # empty stack to store unmatched opening brackets
    for i in s:            # loop for each character in the string                          
        if i in "([{": 
            opening_brackets.append(i)  # open bracket characters are pushed onto the stack
        elif opening_brackets == []:    # if a close bracket character is encountered with nothing in the stack, False
            return False
        elif matches[i] == opening_brackets[-1]: # if a close bracket characters matches the last stacked item, True, and the last stacked item is popped
            opening_brackets.pop()
        else: 
            return False
    return opening_brackets == [] # if the stack is empty at the end, True, False otherwise
    
````


-   Explain why your solution works


````python
# First a dictionary is created that defines all the correct and accepted bracket pairs. 
# Next, an empty stack is created. 
# The for loop iterates once for each character in the inputted string. If the character encountered is a an open bracket it is added to the stack. 
# If the character encountered is a close bracket and the stack is empty, the loop will end and return False. 
# If the character encountered is a close bracket and it matches the last stacked bracket (as defined by the correct matches key value pairs), 
# the last stacked bracket will be popped and the the loop will continue
# If after all iterations have been completed, and the stack is empty, the function will return True because all brackets encountered were successfully matched. 
# If after all iterations have been completed, and the stack is not empty, the function will return False because there are some open brackets that were not closed. 
````


-   Explain the problem’s time and space complexity


````python
# The time complexity is O(n) because there is an iteration for each character in the string. 

# The space complexity is O(n) because, in the worst case scenario, the stack may have to store the entire length of the string. 
# Example: if the string is "((((", the stack will contain four open brackets, "(", which is the entire string. 
````


-   Explain the thinking to an alternative solution (no coding required, but a classmate reading this should be able to code it up based off your text)


````python
# An alternative solution is to search through the string for a character that is beside it's match (i.e. s = "({})", s[1] and s[2] are a match). 
# When a match is found, both characters are removed from the string (i.e s would now be "()"). 
# Removing a matched pair, may result in new matches, which could be removed (i.e. s = "()", s[0] and s[1] are a match). 
# This is continued until the string is empty, which then True is returned. 
# If there are no more matches and the string is not empty, then False is returned. 
````
