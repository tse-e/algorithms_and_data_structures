  Given the `root` of a binary tree, return all root to leaf paths in any order.

  ## Examples

  Input: `root = [1, 2, 2, 3, 5, 6, 7]` *What traversal method is this?*

  ![](./images/q1_ex3.png)

  Input: `root = [10, 9, 7, 8]`

  Output: [[10, 7], [10, 9, 8]]

</details>

#### Starter Code for Question 2

```python
# Definition for a binary tree node.
class TreeNode(object):
  def __init__(self, val = 0, left = None, right = None):
    self.val = val
    self.left = left
    self.right = right

# Function for traversal
def bt_path(root: TreeNode) -> list[list[int]]:
  all_paths = []
  def dfs(node, path):
      if not node:
          return 
      path.append(node.val)
      if not node.left and not node.right:
          all_paths.append(path.copy())
      else:
          dfs(node.left, path)
          dfs(node.right, path)
      path.pop()
  dfs(root, [])
  return all_paths
```

```python
from typing import List # Import List from the typing module for type annotations

def first_duplicate(nums: List[int]) -> int:   # Define a function that takes a list of integers and returns an integer
    seen = set() # Create an empty set to keep track of seen numbers
    for num in nums: # Iterate through each number in the input list
        if num in seen: # Check if the current number has been seen before
            return num  # If it has, return it as the first duplicate
        seen.add(num) # Otherwise, add the number to the set of seen numbers
    return -1 # If no duplicates are found, return -1
```

```python
# Their solution works because it starts with an empty set and iterates through the entire list of numbers
# adding any number that has not been seen (i.e. not in the set yet) to the set.  
# When it gets to a number that is already in the set, this indicates that the number is a duplicate, 
# and their function immediately returns that number, as it is the first instance of a duplicate. 
# If the entire list has been iterated and there are no duplicates, their function returns -1. 
```
